```python
# 필요한 모듈 불러오기
import numpy as np
from scipy.signal import convolve2d # 2D convolution 연산
from skimage.io import imread, imshow # 데이터를 이미지로 보여주기
from skimage.color import rgb2gray # RGB 채널값을 가진 이미지 데이터를, 흑백 이미지의 데이터로 변환하기
```

```python
cat = rgb2gray(imread('cat.jpg')) # read image
```

```python
cat.shape
```

* 흑백으로 이루어진 1channel 이미지

```python
imshow(cat)
```

#### filter 적용하기
* fliter를 적용시킨다는 것은 그저 숫자덩어리에 불과한 커널을 원본이미지에 convolution연산을 적용한다는것
* 이미지 안에 숨어있던 다양하고 복잡한 패턴을 다양한 filter로 추출가능

```python
# Edge Detection
edge = np.array([[-1, -1, -1],
                    [-1, 8, -1],
                    [-1, -1, -1]])
# Sharpen
sharpen = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]])
# Gaussian Blur
blur = (1 / 16.0) * np.array([[1., 2., 1.],
                                 [2., 4., 2.],
                                 [1., 2., 1.]])
```

```python
edge_im = convolve2d(cat, edge, 'valid')
```

```python
imshow(edge_im)
```

```python
sharp_im = convolve2d(cat, sharpen, 'valid')
```

```python
imshow(sharp_im)
```

```python
blur_im = convolve2d(cat,blur,'valid')
```

```python
imshow(blur_im)
```
