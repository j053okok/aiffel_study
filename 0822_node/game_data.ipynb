```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales = pd.read_csv("Sales.csv")
```

```python
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(context='notebook',
              style='darkgrid',
              palette='pastel',
              font='Malgun Gothic',  
              rc=custom_params)
```

```python
sales.info()
```

```python
sales.head()
```

```python
sales["Payment"].sum()
```

```python
sns.histplot(data=sales, x="Payment")
plt.title("결제 현황")
plt.xlabel("결제 금액")
plt.show()
```

```python
# 결제 금액이 0원이 아닌것만 골라내기
sales_1 = sales[sales["Payment"] != 0]
sns.histplot(data=sales_1)
plt.title("무과금 제외 결제현황")
plt.xlabel("결제 금액")
plt.show()
```

```python
sales_1["Payment"].describe()
```

```python
sns.boxplot(data=sales_1["Payment"])
plt.title("무과금 제외 결제현황 박스그래프")
plt.show()
```
