데이터를 확인해보면 1월결제금액은 없음

```python
temp_data = saleslog["Payment"].groupby(saleslog["Month"]).sum()
temp_data
```

```python
ax = plt.bar(temp_data.index, temp_data)
plt.bar_label(ax, labels=[f"{i:,}" for i in temp_data], label_type="center") # 막대그래프에 라벨 입력방법
plt.title("월별 결제량")
plt.show()
```

```python
customerlog = pd.read_csv("customer.csv")
customerlog.head()
```

```python
customerlog.info()
```

```python
piedata = customerlog.groupby(['Gender']).count().Name
plt.pie(piedata, labels=piedata.index, autopct="%1.1f%%")
plt.title("고객 성별 비율")
plt.show()
```

```python
piedata = customerlog.groupby(['OS']).count().Name
plt.pie(piedata, labels=piedata.index, autopct="%1.1f%%")
plt.title("고객 기종 비율")
plt.show()
```

```python
piedata = customerlog.groupby(['MobileCompany']).count().Name
plt.pie(piedata, labels=piedata.index, autopct="%1.1f%%")
plt.title("고객 통신사 비율")
plt.show()
```

```python
# 무과금 유저 제외하고 파이그래프 그리기
total_data = pd.merge(customerlog, saleslog)
paymentcustomer = total_data[total_data["Payment"] != 0]
piedata = paymentcustomer.groupby(["Gender"]).count().Name
plt.pie(piedata, labels=piedata.index, autopct="%1.1f%%")
plt.show()
```

```python
pie_data = paymentcustomer.groupby(['MobileCompany']).count().Name
plt.pie(pie_data, labels=pie_data.index, autopct = '%1.1f%%')
plt.title('Payment Customer Mobile')
plt.show()
```

```python
pie_data = paymentcustomer.groupby(['OS']).count().Name
plt.pie(pie_data, labels=pie_data.index, autopct = '%1.1f%%')
plt.title('Payment Customer OS')
plt.show()
```

```python
# 결제금액기준 월과 성별에 따라 묶기
paymentcustomer['Payment'].groupby([paymentcustomer['Month'], paymentcustomer['Gender']]).sum()
```

```python
male = paymentcustomer[paymentcustomer["Gender"] == "male"]
female = paymentcustomer[paymentcustomer["Gender"] == "female"]
```

```python
temp_data = male['Payment'].groupby(male['Month']).sum()
ax = plt.bar(temp_data.index, temp_data)
plt.bar_label(ax, labels=[f'{x:,.0f}' for x in ax.datavalues], label_type='center')
plt.title('Payments of Male by Month')
plt.show()
```

```python
temp_data = female['Payment'].groupby(female['Month']).sum()
ax = plt.bar(temp_data.index, temp_data)
plt.bar_label(ax, labels=[f'{x:,}' for x in ax.datavalues], label_type='center')
plt.title('Payments of Female by Month')
plt.show()
```
