```python
class DataContainer:

    def __init__(self, x, y):
        self._x = x
        self._y = y

    def plot(self, ax=None, **kwargs):
        if ax is None:
            ax = plt.gca()
        ax.plot(self._x, self._y, **kwargs)
        ax.set_title("Plotted from DataClass")
        return ax
```

```python
data = DataContainer([0, 1, 2, 3, 4], [0, 0.2, 0.5, 0.3, 0.25])
data.plot()
```

### 주석 달기

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
price.plot(ax=ax, style='black')
plt.ylim([130, 150])
plt.xlim(['2022-08-15', '2023-08-09'])

important_data = [(datetime(2022, 8, 15), "Low Price"), (datetime(2023, 8, 9), "Peak Price")]
for d, label in important_data:
    ax.annotate(label, xy=(d, price.asof(d)+10),  # 주석 좌표 (x,y)
                xytext=(d, price.asof(d)+100),  # 주석 텍스트가 위치할 좌표(x,y)
                arrowprops=dict(facecolor='red'))  # 화살표 추가 및 색 설정

ax.set_title('StockPrice')
plt.grid()

plt.show()
```

```python
x = np.linspace(0, 10, 100)  # 0에서 10까지 100개의 숫자를 균등한 간격으로 100개의 숫자를 생성
x
```

```python
# figure 와 add_subplot 을 생략
plt.plot(x, np.sin(x), 'o')
plt.plot(x, np.cos(x), '--', color='red')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x, np.sin(x), '--', color='magenta')

plt.subplot(2,1,2)
plt.plot(x, np.cos(x), 'lime')
plt.show()
```

```python
# linestyle 예시들
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r'); # dotted red
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

```python
plt.plot(x, np.log(x), linestyle='-.', color='m')
plt.show()
```
